```python
import re

month_dict = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}

def rename_my_file(old_file_name: str):
    parts = re.split("-|_", old_file_name)
    parts[0] = month_dict[parts[0]]
    return "-".join(parts)

print(rename_my_file("oct-08_singapore.txt"))
```

```python
for file_name in ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']:
    print(f"{file_name} ----> {rename_my_file(file_name)}")
```

```python
from pathlib import Path

data_dir = Path("os-collaboration-exercise-data")
file_paths = data_dir.glob("*.txt")

for file_path in file_paths:
    # file_path.rename()
    file_path.rename(Path(data_dir/rename_my_file(file_path.stem)).with_suffix(".txt"))
```

```python
folders = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for folder in folders:
    Path(data_dir/folder).mkdir()
```

```python
import shutil

for file_path in data_dir.glob(f"*cairo*", case_sensitive=True):
    shutil.move(file_path, data_dir/"Cairo"/file_path.name)
```

```python
def move_country_files(country: str):
    for file_path in data_dir.glob(f"*{country.lower()}*", case_sensitive=True):
        shutil.move(file_path, data_dir/country.capitalize()/file_path.name)

for folder in folders:
    move_country_files(folder)
```
