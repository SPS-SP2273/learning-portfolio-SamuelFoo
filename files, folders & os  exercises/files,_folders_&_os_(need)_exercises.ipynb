```python
import re

month_dict = {
    "jan": "01",
    "feb": "02",
    "mar": "03",
    "apr": "04",
    "may": "05",
    "jun": "06",
    "jul": "07",
    "aug": "08",
    "sep": "09",
    "oct": "10",
    "nov": "11",
    "dec": "12",
}


def rename_my_file(old_file_name: str):
    parts = re.split("-|_", old_file_name)
    parts[0] = month_dict[parts[0]]
    return "-".join(parts)


print(rename_my_file("oct-08_singapore.txt"))
```

```python
for file_name in [
    "oct-08_singapore.txt",
    "jul-10_cairo.txt",
    "may-15_dhaka.txt",
    "may-13_cairo.txt",
    "oct-21_cairo.txt",
    "jan-10_singapore.txt",
    "jun-20_tokyo.txt",
    "aug-06_jakarta.txt",
    "dec-21_karachi.txt",
    "jan-01_tokyo.txt",
]:
    print(f"{file_name} ----> {rename_my_file(file_name)}")
```

```python
from pathlib import Path

data_dir = Path("os-collaboration-exercise-data")
file_paths = data_dir.glob("*.txt")

for file_path in file_paths:
    new_name = Path(data_dir / rename_my_file(file_path.stem)).with_suffix(".txt")
    file_path.rename(new_name)
    print(f"Renamed {file_path}".ljust(70) + f"as {new_name}")
```

```python
folders = [
    "Cairo",
    "Dhaka",
    "Jakarta",
    "Karachi",
    "Manila",
    "Melbourne",
    "Osaka",
    "Shanghai",
    "Singapore",
    "Tokyo",
]


for folder in folders:
    new_dir = data_dir / folder

    if new_dir.exists():
        print(f"{str(new_dir).ljust(50)} already exists. Skipping...")
    else:
        Path(new_dir).mkdir(parents=True)
        print(f"Created directory \t {new_dir}")
```

```python
import shutil

for file_path in data_dir.glob("*cairo*", case_sensitive=True):
    new_path = data_dir / "*cairo*".capitalize() / file_path.name
    shutil.move(file_path, data_dir / "Cairo" / file_path.name)
    print(f"Moved {file_path}".ljust(70) + f"to {new_path}")
```

```python
def move_country_files(country: str):
    for file_path in data_dir.glob(f"*{country.lower()}*", case_sensitive=True):
        print(
            f"Moving {file_path}".ljust(70)
            + f"to {data_dir / country.capitalize() / file_path.name}"
        )
        shutil.move(file_path, data_dir / country.capitalize() / file_path.name)


for folder in folders:
    move_country_files(folder)
```
