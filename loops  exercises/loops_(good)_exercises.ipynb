# Exercise 3

```python
%%timeit
lst = []
for i in range(1, 10, 2):
    lst.append(i**2)
lst
```

```python
%%timeit
[i**2 for i in range(1, 10, 2)]
```

List comprehension is faster. This is because suspending function frames in a stack runs faster than loading the attribute append of the list and calling the function.

# Exercise 5

```python
co2_data = [
    (2000, 369.55),
    (2001, 371.14),
    (2002, 373.28),
    (2003, 375.80),
    (2004, 377.52),
    (2005, 379.80),
    (2006, 381.90),
    (2007, 383.79),
    (2008, 385.60),
    (2009, 387.43),
    (2010, 389.90),
    (2011, 391.65),
    (2012, 393.85),
    (2013, 396.52),
    (2014, 398.65),
    (2015, 400.83),
    (2016, 404.24),
    (2017, 406.55),
    (2018, 408.52),
    (2019, 411.44),
    (2020, 414.24),
]
```

```python
for i, (year, co2) in enumerate(co2_data):
    if i == 0:
        continue
    change = co2 - co2_data[i - 1][1]
    if change >= 3:
        print(f"Year: {round(year)}, Change: +{round(change, 2)}.")
```

```python
import numpy as np

co2_data = np.array(co2_data)
# Get CO2 level changes, prepending a change of 0 for year 2000
co2_changes = np.diff(co2_data[:, 1], prepend=co2_data[0, 1])
# Stack the CO2 level changes as the third column
data = np.hstack([co2_data, co2_changes[:, np.newaxis]])
# Filter
filtered_data = data[co2_changes >= 3]

for p in filtered_data:
    print(f"Year: {round(p[0])}, Change: +{round(p[2], 2)}")
```
