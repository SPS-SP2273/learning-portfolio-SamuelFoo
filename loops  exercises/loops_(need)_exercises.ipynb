# Exercise 1

```python
temperatures_celsius = [
    0, 5, 10, 15, 20, 25,
    30, 35, 40, 45, 50
]

def fahrenheit(celsius):
    return celsius * 9 / 5 + 32

temperatures_fahrenheit = list(map(fahrenheit, temperatures_celsius))

print("Celsius\tFahrenheit")
for temperature_celsius, temperature_fahrenheit in zip(temperatures_celsius, temperatures_fahrenheit):
    print(f"{temperature_celsius}\t{temperature_fahrenheit}")
```

# Exercise 2
Write a Python snippet that prints a multiplication table (up to 5) for numbers 1 through 5 using nested for loops. The output should be formatted as shown below:


```python
print_str = ""
for i in range(1, 6):
    print_str += f"{i} : "
    for j in range(1, 6):
        print_str += f"{i*j}\t"
    print_str += "\n"
print(print_str.expandtabs(4))
```
