# 2 Lists vs. Arrays

## 2.1 Size

```python
%code_wrap
py_list_2d = [[1, "A"], [2, "B"], [3, "C"], [4, "D"],
              [5, "E"], [6, "F"], [7, "G"], [8, "H"],
              [9, "I"], [10, "J"]]

np_array_2d = np.array(py_list_2d)      # Reusing the Python list 
                                        # to create a NEW
                                        # NumPy array
```

```python
len(py_list_2d)
len(np_array_2d)
np_array_2d.shape
```

## 2.2 Arrays are fussy about type

```python
py_list = [1, 1.5, "A"]
np_array = np.array(py_list)
```

```python
py_list
```

```python
np_array
```

## 2.3 Adding a number

```python
%code_wrap
py_list = [1, 2, 3, 4, 5]
np_array = np.array(py_list)         # Reusing the Python list
                                     # to create a NEW
                                     # NumPy array
```

```python
np_array + 10
```

```python
# py_list + 10        # Won't work!
```

## 2.4 Adding another list

```python
py_list_1 = [1, 2, 3, 4, 5]
py_list_2 = [10, 20, 30, 40, 50]

np_array_1 = np.array(py_list_1)
np_array_2 = np.array(py_list_2)
```

```python
py_list_1 + py_list_2
```

```python
np_array_1 + np_array_2
```

## 2.5 Multiplying by a Number

```python
py_list = [1, 2, 3, 4, 5]
np_array = np.array(py_list)
```

```python
py_list * 2
```

```python
np_array * 2
```

## 2.6 Squaring

```python
py_list = [1, 2, 3, 4, 5]
np_array = np.array(py_list)
```

```python
np_array**2
```

```python
# py_list**2                      # Won't work!
```

## 2.7 Asking questions

```python
py_list = [1, 2, 3, 4, 5]
np_array = np.array(py_list)
```

```python
py_list == 3  # False because different objects
```

```python
np_array == 3
```

```python
# py_list > 3      # Won't work!
```

```python
np_array > 3
```

## 2.8 Mathematics

```python
py_list = [1, 2, 3, 4, 5]
np_array = np.array(py_list)
```

```python
sum(py_list)  # sum() is a base Python function
```

```python
max(py_list)  # max() is a base Python function
```

```python
min(py_list)  # min() is a base Python function
```

```python
np_array.sum()
```

```python
np_array.max()
```

```python
np_array.min()
```

```python
np_array.mean()
```

```python
np_array.std()
```
