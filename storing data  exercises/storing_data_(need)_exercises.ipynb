# Exercise 1

1. Two similarities between lists and arrays.
- Both are data structures.
- Indexing has a time complexity of O(1). Space complexity is O(N) where N is the number of elements.

2. Two differences between lists and arrays.
- Lists can contain objects of different types. Arrays can only contain objects of the same type.
- NumPy arrays support vectorisation.

3. What is a dictionary?
- Implemented by a hashmap, where each object is associated with a unique hash key generated by a hash function. This makes finding/ checking whether an element exist run in O(1) time. However, the keys must be hashable (e.g. lists can't work as keys but integers can.)

# Exercise 2

```python
py_list = ["a1", "b2", "c3", "d4", "e5", "f6", "g7", "h8", "i9", "j10"]
print(*list(filter(lambda x: int(x[1:]) % 2, py_list)), sep="\n")
```

# Exercise 3

```python
elements = ['Hydrogen',
            'Helium', 'Lithium',
            'Beryllium', 'Boron', 'Carbon',
            'Nitrogen', 'Oxygen',
            'Fluorine',
            'Neon']
```

```python
# Access and print the element at index 4 using forward indexing.
print(elements[4])

# Access and print the element at index 4 from the end of the list using reverse indexing.
print(elements[4-len(elements)])
```
