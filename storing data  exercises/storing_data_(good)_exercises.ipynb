# Exercise 2

```python
import numpy as np
```

```python
np_array_2d = np.array(
    [
        [1, "A"],
        [3, "C"],
        [2, "B"],
        [4, "D"],
        [5, "E"],
        [7, "G"],
        [6, "F"],
        [8, "H"],
        [10, "J"],
        [9, "I"],
    ]
)
```

```python
result = np_array_2d[np_array_2d[:, 0].astype(np.int64) % 2 == 1][:, 1]
result
```

```python
# Optionally, print in the given format
print("[" + ", ".join(result) + "]")
```
