```python
from pandas import Interval
import sys


class Switch(dict):
    def __getitem__(self, item):
        for key in self.keys():
            if item in key:
                return super().__getitem__(key)
        raise KeyError


grade_switch = Switch(
    {
        Interval(0, 34, closed="both"): "Fail",
        Interval(35, 49, closed="both"): "C",
        Interval(50, 69, closed="both"): "B",
        Interval(70, 100, closed="both"): "A",
    }
)
```

```python
grade = input("Enter the student's score:")
print(f"Enter the student's score: {grade}")
print(f"The student's letter grade is: {grade_switch[int(grade)]}")
```
