```python
from enum import Enum


# To reduce the occurrence of magic numbers
class TemperatureUnit(Enum):
    CELSIUS = "Celsius"
    FAHRENHEIT = "Fahrenheit"
    KELVIN = "Kelvin"
```

# Exercise 1

```python
def convert_celsius(temperature_celsius, target_scale=TemperatureUnit.FAHRENHEIT.value):
    if TemperatureUnit(target_scale) == TemperatureUnit.FAHRENHEIT:
        return 9 / 5 * temperature_celsius + 32
    elif TemperatureUnit(target_scale) == TemperatureUnit.KELVIN:
        return temperature_celsius + 273.15
    elif TemperatureUnit(target_scale) == TemperatureUnit.CELSIUS:
        return temperature_celsius


print(convert_celsius(25))
print(convert_celsius(25, target_scale="Kelvin"))
```

# Exercise 2

```python
def convert_fahrenheit(
    temperature_fahrenheit, target_scale=TemperatureUnit.CELSIUS.value
):
    if TemperatureUnit(target_scale) == TemperatureUnit.CELSIUS:
        return (temperature_fahrenheit - 32) / 1.8
    elif TemperatureUnit(target_scale) == TemperatureUnit.KELVIN:
        return (temperature_fahrenheit - 32) * 5 / 9 + 273.15
    elif TemperatureUnit(target_scale) == TemperatureUnit.FAHRENHEIT:
        return temperature_fahrenheit


print(convert_fahrenheit(32))
print(convert_fahrenheit(32, "Kelvin"))
```

# Exercise 3

```python
def convert_temperature(temperature, source_scale, target_scale):
    if TemperatureUnit(source_scale) == TemperatureUnit.CELSIUS:
        return convert_celsius(temperature, target_scale)
    elif TemperatureUnit(source_scale) == TemperatureUnit.FAHRENHEIT:
        return convert_fahrenheit(temperature, target_scale)
    elif TemperatureUnit(source_scale) == TemperatureUnit.KELVIN:
        temperature_celsius = temperature - 273.15
        return convert_celsius(temperature_celsius, target_scale)


print(convert_temperature(0, "Kelvin", "Celsius"))
```
