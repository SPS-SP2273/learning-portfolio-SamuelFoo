# Exercise 1

````python
def greeting(name):
    if name == 'Batman':
        return 'Hello Batman! So, nice to meet you!'
    return f'Hello {name}!'
````

If the name is ```"Batman"```, the line ```'Hello Batman! So, nice to meet you!'``` will run. If the name is not ```"Batman"```, the ```if``` block will not run and the line ```return f'Hello {name}!'``` will run instead.

# Exercise 2

````python
import sys
from pandas import Interval

class Switch(dict):
    def __getitem__(self, item):
        for key in self.keys():
            if item in key:
                return super().__getitem__(key)
        raise KeyError

# Modified intervals slightly to remove ill-defined points
bmi_switch = Switch({
    Interval(0, 18.5, closed="left"): "Underweight",
    Interval(18.5, 25, closed="left"): "Normal weight",
    Interval(25, 30, closed="left"): "Overweight",
    Interval(30, sys.maxsize, closed="left"): "Obesity"
})

def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi_switch[bmi]

print(calculate_bmi(16, 1))
print(calculate_bmi(18.5, 1))
print(calculate_bmi(25, 1))
print(calculate_bmi(30, 1))
````
